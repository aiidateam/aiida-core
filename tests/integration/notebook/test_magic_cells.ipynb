# Test 2: Magic cells after load_profile

This tests that `%%bash`, `!echo`, `%who` etc. work after the greenback patch is installed.

```python
%%bash
echo "Hello from bash!"
echo "This should work fine."
date
```

```python
from aiida import load_profile
from aiida.storage.sqlite_temp import SqliteTempBackend

profile = load_profile(
    SqliteTempBackend.create_profile(
        'test-magic',
        options={'warnings.development_version': False, 'runner.poll.interval': 1},
        debug=False,
    ),
    allow_switch=True,
)
print('Profile loaded:', profile)
```
