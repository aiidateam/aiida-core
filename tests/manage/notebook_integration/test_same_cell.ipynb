# Test 1: load_profile + multiply in separate cells

Tests that `load_profile()` in the same cell while having an engine call in the same cells, raises a proper explanation.

```python
from aiida import engine, load_profile, orm
from aiida.storage.sqlite_temp import SqliteTempBackend

profile = load_profile(
    SqliteTempBackend.create_profile(
        'test-same-cell',
        options={'warnings.development_version': False, 'runner.poll.interval': 1},
        debug=False,
    ),
    allow_switch=True,
)
print('Profile loaded')


@engine.calcfunction
def multiply(x, y):
    return x * y


# This works because load_profile() was called in a prior cell:
result = multiply(orm.Int(3), orm.Int(4))
print(f'3 * 4 = {result}')
assert result == 12, f'Expected 12, got {result}'

# Verify it also works for subsequent calls
result2 = multiply(orm.Int(5), orm.Int(6))
print(f'5 * 6 = {result2}')
assert result2 == 30
```
